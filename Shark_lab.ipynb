```python
# function checking the Frequency of the different elements in each column of a dataframe
def freq_element_percolumn(data):
    cols=data.columns
    for i in cols:
        freq_per_element=pd.Series(data[i]).value_counts(normalize=True)*100
        print(f'Column studied is : {i}')
        display(freq_per_element)
    return

freq_element_percolumn(initial_data)
```

# CLEANING

```python
#Deleting not necessary columns. Input dataframe,liste of columns to delete
def suppress_columns(data,cols):
    newdata=data.copy()
    newdata=newdata.drop_duplicates()
    newdata=newdata.drop(cols,axis=1)
    return newdata

new_data=suppress_columns(initial_data,['Unnamed: 22','Unnamed: 23'])
display(new_data)

```
